Summarize code example

```python

code = "print('hello world!')"

inputs = tokenizer.encode("summarize: " + code, return_tensors="pt", max_length=512, truncation=True)

outputs = model.generate(inputs, max_length=100, num_beams=3, early_stopping=True)

summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Summarized code: " + summary)
```
